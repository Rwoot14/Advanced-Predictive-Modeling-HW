Instead of recording the time for each lead, the time taken to generate the next 5 leads are recorded, i.e., there is one recording (denoting the elapsed time) for every 5 consecutive leads. $$ p(t) = \frac{1}{C \times \theta^{5}}t^{4}exp^{-\frac{t}{\theta}},$$
(where C is a normalizing constant). Therefore, if $\theta$ can be estimated, then he can provide detailed information
about the lead generation process, including average rates, variances etc.

## Answer

For the sweep of the regularization parameter, we will look at a grid of values ranging from $\lambda = 10^{10}$ to $\lambda = 10^{-2}$. In Python, you can consider this range of values as follows:

      import numpy as np

```python
from sklearn import datasets
from sklearn.linear_model import LassoCV
from sklearn.linear_model import Lasso
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

Before we fit lasso or ridge models to the data, it is a good practice to standardize the features. This is because Lasso and Ridge calculate the cost function and hence lambda based on the coefficient values of each independent variable. Because the coefficient values highly depend on the magnitude of each variable, it is necessary to standardize so that the cost function is not affected.

```python
from sklearn.preprocessing import scale
X_train_scaled = scale(X_train, axis=0, with_mean=True, with_std=True, copy=True)
X_test_scaled = scale(X_test, axis=0, with_mean=True, with_std=True, copy=True)
```

```python
# Lasso setup for cross validating the shrinkage parameter with K_fold cv.
lasso = Lasso(random_state=0)
alphas = 10**np.linspace(10,-2,100)*0.5

tuned_parameters = [{'alpha': alphas}]
n_folds = 5

# Uses function gridsearch to iterate the lasso model over the k_folds
clf = GridSearchCV(lasso, tuned_parameters, cv=n_folds, refit=False,scoring='neg_mean_squared_error')#refit needs to be true?? need to standardize x's??
clf.fit(X_train_scaled, y_train)
#stores the scores of the cross validation
scores = clf.cv_results_['mean_test_score']
scores_std = clf.cv_results_['std_test_score']
```

```python
lasso_param=clf.best_params_
print(clf.best_params_)
```

The optimal lambda for the cross validated lasso regression is the alpha show above. The lambda is relatively small meaning that the model needs to be more complex to get the optimal MSE.  

```python
from sklearn.linear_model import Ridge
# Ridge Regression
ridge = Ridge(random_state=0)
alphas = 10**np.linspace(10,-2,100)*0.5

tuned_parameters = [{'alpha': alphas}]
n_folds = 5

clf = GridSearchCV(ridge, tuned_parameters, cv=n_folds, refit=False,scoring='neg_mean_squared_error')
clf.fit(X_train_scaled, y_train)
scores = clf.cv_results_['mean_test_score']
scores_std = clf.cv_results_['std_test_score']
```

```python
ridge_param=clf.best_params_
print(clf.best_params_)
```

The optimal lambda for the cross validated Ridge regression is the alpha shown above. The lambda ismuch larger than the lasso regression. This is likley do to the inability for the ridge regression to force the coef.  to zero.

As alpha increases, more and more coefficients in the model become zero i.e the model becomes simpler and simpler. This means that the variance in the model goes down whereas bias increases. The ideal value of lambda is when the model is not too complex but also not too simple. Unlike in the case of ridge where the cost function involves a squared term and hence the coefficients take a long time to converge to zero exactly, it is possible for lasso to force the coefficients to zero as seen from the plot.

```python
ridge = Ridge(random_state=0)
alphas = 10**np.linspace(10,-2,100)*0.5
tuned_parameters = [{'alpha': alphas}]
n_folds = 5
coefs = []
for a in alphas:
    ridge.set_params(alpha=a)
    ridge.fit(X_train_scaled, y_train)
    coefs.append(ridge.coef_)


ax = plt.gca()
ax.plot(alphas*2, coefs)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
```

We can see that the coefficients tend to become close to zero, but it takes very very high values of lambda to force the coefficients to zero. This is because ridge regression has squared coefficient terms as penalty term in the cost function against the first order penalty term in lasso.

# Question 5  (5 pts)

A regression model that includes "interaction terms" (i.e. quadratic terms of the form $x_ix_j$) as predictors in addition to the linear terms is clearly more general than a corresponding model that employs the same independent variables but only uses the linear terms. Outline two situations where the simpler (less general) model would be preferred to the more powerful model that includes interactive terms.

## Answer



1. In situations where the number of training data points available is very less, trying to fit a complex general model with interaction terms can result in a high variance in the model. Such a model will perform badly on a test dataset when compared to the simple linear model. It is preferable to have a simpler model in this case.
2. Complex model does not always imply better model. If the added complexity in the model does not give any significant improvement in the prediction accuracy, it's better to use the simpler model. Adding interaction terms to the model can take away interpretability from the model. Therefore, simpler is better unless the additional complexity results in a better predictive capability.
